```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tabulate import tabulate

# Mengidentifikasi kolom-kolom kategorikal
categorical_columns = data.select_dtypes(include=['object']).columns

# Mengubah kolom-kolom kategorikal menjadi numerik
for column in categorical_columns:
    label_encoder = LabelEncoder()
    data[column] = label_encoder.fit_transform(data[column])

# Memisahkan fitur dan target
X = data.drop('Attack_type', axis=1)
y = data['Attack_type']

# Membagi data menjadi set pelatihan dan set pengujian secara proporsional untuk setiap kategori
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Menstandarkan fitur
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Mengonversi hasil standarisasi kembali ke DataFrame untuk kemudahan tampilan
X_train_df = pd.DataFrame(X_train, columns=X.columns)
X_test_df = pd.DataFrame(X_test, columns=X.columns)
y_train_df = pd.DataFrame(y_train.reset_index(drop=True), columns=['Attack_type'])
y_test_df = pd.DataFrame(y_test.reset_index(drop=True), columns=['Attack_type'])

# Menampilkan tabel train setelah distandarisasi
print("Tabel Train Setelah Distandarisasi:")
print(tabulate(X_train_df.head(), headers="keys", tablefmt="pretty"))

# Menampilkan tabel test setelah distandarisasi
print("\nTabel Test Setelah Distandarisasi:")
print(tabulate(X_test_df.head(), headers="keys", tablefmt="pretty"))

# Menampilkan target train
print("\nTarget Train:")
print(tabulate(y_train_df.head(), headers="keys", tablefmt="pretty"))

# Menampilkan target test
print("\nTarget Test:")
print(tabulate(y_test_df.head(), headers="keys", tablefmt="pretty"))
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tabulate import tabulate

# Melatih model Random Forest
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=0
)

# Train the model
rf_model.fit(X_train, y_train)

# Memprediksi dan mengevaluasi model
y_pred_rf = rf_model.predict(X_test)

# Akurasi model keseluruhan
accuracy = accuracy_score(y_test, y_pred_rf)
accuracy_percentage = accuracy * 100
print(f"Random Forest Accuracy: {accuracy_percentage:.2f}%")
```

```python

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred_rf)
print("\nConfusion Matrix:")
print(conf_matrix)

# Fungsi untuk menghitung TP, FP, FN, TN dari confusion matrix
def calculate_tp_fp_fn_tn(conf_matrix):
    tp = conf_matrix.diagonal()
    fp = conf_matrix.sum(axis=0) - tp
    fn = conf_matrix.sum(axis=1) - tp
    tn = conf_matrix.sum() - (tp + fp + fn)
    return tp, fp, fn, tn

tp_rf, fp_rf, fn_rf, tn_rf = calculate_tp_fp_fn_tn(conf_matrix)

# Mengonversi TP, FP, FN, TN menjadi DataFrame
tp_df = pd.DataFrame({
    'Category': rf_model.classes_,
    'TP': tp_rf
})

fp_df = pd.DataFrame({
    'Category': rf_model.classes_,
    'FP': fp_rf
})

fn_df = pd.DataFrame({
    'Category': rf_model.classes_,
    'FN': fn_rf
})

tn_df = pd.DataFrame({
    'Category': rf_model.classes_,
    'TN': tn_rf
})

# Menampilkan tabel TP, FP, FN, dan TN
print("\nTrue Positives (TP) per Category:")
print(tabulate(tp_df, headers='keys', tablefmt='pretty'))

print("\nFalse Positives (FP) per Category:")
print(tabulate(fp_df, headers='keys', tablefmt='pretty'))

print("\nFalse Negatives (FN) per Category:")
print(tabulate(fn_df, headers='keys', tablefmt='pretty'))

print("\nTrue Negatives (TN) per Category:")
print(tabulate(tn_df, headers='keys', tablefmt='pretty'))

# Mengubah target names menjadi string
target_names = [str(i) for i in rf_model.classes_]

# Classification report
classification_report_str = classification_report(y_test, y_pred_rf, target_names=target_names)
print("\nClassification Report:")
print(classification_report_str)

# Mengubah classification report menjadi DataFrame
classification_report_dict = classification_report(y_test, y_pred_rf, target_names=target_names, output_dict=True)
classification_report_df = pd.DataFrame(classification_report_dict).transpose()

# Extract akurasi per kategori dari classification report
classification_report_dict = classification_report(y_test, y_pred_rf, target_names=target_names, output_dict=True)
category_accuracy = {category: metrics['recall'] for category, metrics in classification_report_dict.items() if category != 'accuracy'}

# Tampilkan akurasi per kategori
category_accuracy_df = pd.DataFrame(list(category_accuracy.items()), columns=['Category', 'Accuracy'])
category_accuracy_df['Accuracy'] = (category_accuracy_df['Accuracy'] * 100).round(2)  # Convert to percentage and round to 2 decimal places
print("\nAccuracy per Category:")
print(tabulate(category_accuracy_df, headers='keys', tablefmt='pretty'))

# Menyimpan hasil ke Excel dan file teks
results_excel_path = '/content/drive/MyDrive/Hasil Algoritma/Random Forest/random_forest_results.xlsx'
classification_report_excel_path = '/content/drive/MyDrive/Hasil Algoritma/Random Forest/random_forest_classification_report.xlsx'
conf_matrix_excel_path = '/content/drive/MyDrive/Hasil Algoritma/Random Forest/confusion_matrix.xlsx'
category_accuracy_excel_path = '/content/drive/MyDrive/Hasil Algoritma/Random Forest/category_accuracy.xlsx'
tp_excel_path = '/content/drive/MyDrive/Hasil Algoritma/Random Forest/tp.xlsx'
fp_excel_path = '/content/drive/MyDrive/Hasil Algoritma/Random Forest/fp.xlsx'
fn_excel_path = '/content/drive/MyDrive/Hasil Algoritma/Random Forest/fn.xlsx'
tn_excel_path = '/content/drive/MyDrive/Hasil Algoritma/Random Forest/tn.xlsx'

# Simpan hasil prediksi dan evaluasi
results_df = pd.DataFrame({"Actual": y_test.reset_index(drop=True), "Predicted": y_pred_rf})
results_df.to_excel(results_excel_path, index=False)

# Simpan confusion matrix ke Excel
conf_matrix_df = pd.DataFrame(conf_matrix, index=target_names, columns=target_names)
conf_matrix_df.to_excel(conf_matrix_excel_path)

# Simpan classification report ke Excel
classification_report_df.to_excel(classification_report_excel_path, index=True)

# Simpan akurasi per kategori ke Excel
category_accuracy_df.to_excel(category_accuracy_excel_path, index=False)

# Simpan TP, FP, FN, dan TN ke Excel
tp_df.to_excel(tp_excel_path, index=False)
fp_df.to_excel(fp_excel_path, index=False)
fn_df.to_excel(fn_excel_path, index=False)
tn_df.to_excel(tn_excel_path, index=False)

# Cetak informasi path penyimpanan
print(f"\nHasil prediksi Random Forest telah disimpan di: {results_excel_path}")
print(f"Confusion Matrix telah disimpan di: {conf_matrix_excel_path}")
print(f"Classification Report telah disimpan di: {classification_report_excel_path}")
print(f"Accuracy per Category telah disimpan di: {category_accuracy_excel_path}")
print(f"True Positives (TP) telah disimpan di: {tp_excel_path}")
print(f"False Positives (FP) telah disimpan di: {fp_excel_path}")
print(f"False Negatives (FN) telah disimpan di: {fn_excel_path}")
print(f"True Negatives (TN) telah disimpan di: {tn_excel_path}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv1D, MaxPooling1D, Flatten
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Membentuk ulang data untuk CNN
X_train_cnn = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test_cnn = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Mengidentifikasi jumlah kelas
num_classes = len(label_encoder.classes_)

# Membuat model CNN
cnn_model = Sequential()
cnn_model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(X_train_cnn.shape[1], 1)))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Flatten())
cnn_model.add(Dense(100, activation='relu'))
cnn_model.add(Dropout(0.5))
cnn_model.add(Dense(num_classes, activation='softmax'))  # Menggunakan num_classes sebagai jumlah unit di layer output

# Kompilasi model
cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Melatih model
cnn_model.fit(X_train_cnn, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Memprediksi dan mengevaluasi model
y_pred_proba = cnn_model.predict(X_test_cnn)  # Menghasilkan probabilitas untuk setiap kelas
y_pred_cnn = y_pred_proba.argmax(axis=1)  # Mengambil kelas dengan probabilitas tertinggi

# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred_cnn)
accuracy_percentage = accuracy * 100
print(f"Accuracy: {accuracy_percentage:.2f}%")

# Menyimpan hasil ke Excel dan file teks
results_excel_path = '/content/drive/MyDrive/Hasil Algoritma/CNN/cnn_results.xlsx'
classification_report_path = '/content/drive/MyDrive/Hasil Algoritma/CNN/cnn_classification_report.txt'
conf_matrix_excel_path = '/content/drive/MyDrive/Hasil Algoritma/CNN/cnn_confusion_matrix.xlsx'

# Simpan hasil prediksi dan evaluasi
results_df = pd.DataFrame({"Actual": y_test.reset_index(drop=True), "Predicted": y_pred_cnn})
results_df.to_excel(results_excel_path, index=False)

# Simpan confusion matrix ke Excel
conf_matrix = confusion_matrix(y_test, y_pred_cnn)
conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
conf_matrix_df.to_excel(conf_matrix_excel_path)

# Simpan classification report ke file teks
classification_report_str = classification_report(y_test, y_pred_cnn)
with open(classification_report_path, 'w') as f:
    f.write(classification_report_str)

# Cetak informasi path penyimpanan
print(f"\nHasil prediksi CNN telah disimpan di: {results_excel_path}")
print(f"Classification report telah disimpan di: {classification_report_path}")
print(f"Confusion Matrix telah disimpan di: {conf_matrix_excel_path}")
```
